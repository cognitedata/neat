```python
from cognite.neat._rules.analysis import InformationAnalysis
from cognite.neat._rules.models._types import ClassEntity
```

```python
def get_tables_and_columns(rules):

    columns = {}

    class_properties = InformationAnalysis(rules).classes_with_properties()


    for class_entity, properties in class_properties.items():

        columns[class_entity.suffix] = ["ID"] + [property_.property_ for property_ in properties]

    return columns
```

```python
tables_and_columns = get_tables_and_columns(neat._state.last_verified_information_rules)
```

```python
from copy import deepcopy
```

```python
tables = {}


for table, columns in tables_and_columns.items():

    row_template = {column: None for column in columns}
    tables[table] = []
    for instance in neat._state.store.read(table):

        row = deepcopy(row_template)
        row["ID"] = instance[0]

        for key,value in instance[1].items():
            if str(key) == 'http://www.w3.org/1999/02/22-rdf-syntax-ns#type':
                continue

            row[key] = value

        tables[table].append(row)
```

```python
from collections import OrderedDict

# Create an OrderedDict
ordered_dict = OrderedDict()

# Add items to the OrderedDict
ordered_dict['first'] = 1
ordered_dict['second'] = 2
ordered_dict['third'] = 3

# Example usage
print("OrderedDict contents:")
for key, value in ordered_dict.items():
    print(f"{key}: {value}")

# Example of reordering items
ordered_dict.move_to_end('first')  # Move 'first' to the end
ordered_dict.move_to_end('second', last=False)  # Move 'second' to the beginning

print("\nReordered OrderedDict contents:")
for key, value in ordered_dict.items():
    print(f"{key}: {value}")
```
